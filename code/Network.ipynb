```python
import networkx as nx
import matplotlib.pyplot as plt
import community  # from python-louvain
```

```python
# Load edge list
G = nx.read_edgelist("../data/facebook_combined.txt")
print(f"Nodes: {G.number_of_nodes()}")
print(f"Edges: {G.number_of_edges()}")
print(f"Is directed: {G.is_directed()}")
```

```python
partition = community.best_partition(G)
colors = [partition[node] for node in G.nodes()]
```

```python
num_communities = len(set(partition.values()))
print(f"Number of communities: {num_communities}")
```

```python
#count number of nodes in each community
community_counts = {}
for comm in partition.values():
    if comm in community_counts:
        community_counts[comm] += 1
    else:
        community_counts[comm] = 1
for comm_id, count in sorted(community_counts.items()):
    print(f"Community {comm_id}: {count} nodes")
```

```python
plt.figure(figsize=(12, 12))
nx.draw_spring(
    G,
    node_size=20,
    node_color=colors,
    edge_color="lightgray",
    with_labels=False
)
plt.title("Ego-Facebook Network with Communities")
plt.savefig("../figures/ego_facebook_communities.png")
plt.show()
```
